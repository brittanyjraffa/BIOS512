Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

#Yes, the males tend to be bigger across all species. 

```r
options(repr.plot.width=4, repr.plot.height=5)
p=ggplot(penguins, aes(y=sex, x=body_mass_g)) + geom_point(na.rm = T, size = 1.0, alpha = 0.25, position = position_jitter(height = 0.2, width = 0)) +
facet_wrap(~species, ncol=1) + scale_x_continuous(labels=comma) + 
labs(y='Penguin sex', x='Body mass in grams', title = 'Scatterplot of penguin body mass by sex and species') +
theme_set(theme_minimal(base_size = 10)) +  
theme(plot.title = element_text(size = 8, face="bold", hjust=0.5), axis.title.x=element_text(size=8), axis.title.y=element_text(size=8), axis.text.y=element_text(color=c('pink','blue', 'red')))
p
```
