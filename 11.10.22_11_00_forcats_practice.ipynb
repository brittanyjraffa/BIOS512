```r
chocolate = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-01-18/chocolate.csv', col_types = cols()) 
    
choc2 = chocolate |>
    select(country_of_bean_origin, rating, cocoa_percent)|>
    mutate(country_of_bean_origin = country_of_bean_origin |> 
           fct_lump(n=8) |> fct_infreq() |> fct_rev() |>
            fct_relevel('Other', after = Inf))

p=ggplot(choc2, aes(y=country_of_bean_origin)) + geom_bar()
p
```

```r
#Chuck's example:
volcano = read_csv(
    'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-12/volcano.csv',
    col_types = cols()
) 

volcano2 = volcano |> 
    select(volcano_name, region, last_eruption_year) |>
    mutate(region = region |> 
               fct_lump(n = 12) |> 
               fct_infreq() |> fct_rev() |> 
               fct_relevel('Other'))
    
p = ggplot(volcano2, aes(y = region)) + geom_bar()
           
p
```
