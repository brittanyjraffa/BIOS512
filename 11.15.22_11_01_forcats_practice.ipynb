```r
median_values = mpg |>
    group_by(class) |>
    summarize(cty = median(cty))

p = ggplot(mpg, aes(x = cty, y = class)) +
    geom_point(alpha = 0.35, position = position_jitter(height = 0.2, width = 0)) +
    geom_crossbar(color = 'steelblue', 
                  data = median_values, mapping = aes(xmin = cty, xmax = cty))

p
```

```r
#now we need to turn class into a factor and set the levels in an order using reorder function. Create a new table mpg2. 

median_values = mpg |>
    group_by(class) |>
    summarize(cty = median(cty))

mpg2 = mpg |>
    mutate(class = fct_reorder(class,cty, .fun = median)) #could also pick the max function and then y values sorted by max, .fun=max

p = ggplot(mpg2, aes(x = cty, y = class)) +
    geom_point(alpha = 0.35, position = position_jitter(height = 0.2, width = 0)) +
    geom_crossbar(color = 'steelblue', 
                  data = median_values, mapping = aes(xmin = cty, xmax = cty))

p
```

```r
#another way to do this without using fct_reorder function

mpg |> 
    group_by(class) |>
    mutate(cty_median = median(cty)) |> #this represents the median value of cty within each class
    ungroup() |>
    arrange(cty_median) |>
    mutate(class = fct_inorder(class)) |>
    pull(class) |> levels()
```

```r
median_values = mpg |>
    group_by(class) |>
    summarize(cty = median(cty))

# mpg2 = mpg |>
#     mutate(class = fct_reorder(class, cty, .fun = median))

mpg2 = mpg |>
    group_by(class) |>
    mutate(cty_median = median(cty)) |>
    ungroup() |>
    arrange(cty_median) |>
    mutate(class = fct_inorder(class))

p = ggplot(mpg2, aes(x = cty, y = class)) +
    geom_point(alpha = 0.35, position = position_jitter(height = 0.2, width = 0)) +
    geom_crossbar(color = 'steelblue', 
                  data = median_values, mapping = aes(xmin = cty, xmax = cty))

p
```
